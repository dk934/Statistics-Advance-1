1. What is a random variable in probability theory?

A random variable in probability theory is a numerical value that represents the outcome of a random experiment. It assigns a number to each possible outcome of an event.
   

2. What are the types of random variables?
Random variables are classified into two main types:

Discrete Random Variable
Takes on a finite or countable number of distinct values. It typically represents countable events (e.g., number of students in a class, number of heads in coin flips).

Continuous Random Variable
It takes on an uncountable number of values within an interval. It Often represents measurements (e.g., height, weight, time).

4.What are probability distribution functions?

Probability Distribution Function (PDF) describes how probabilities are distributed over the values of a random variable. There are two main types:

Probability Mass Function:
Used for discrete random variables (e.g., number of heads in coin flips). The sum of probabilities for all possible values must be 1.

Probability Density Function:
Used for continuous random variables (e.g., height, weight, temperature). Unlike PMFs, PDFs do not give probabilities of specific values. Instead, probabilities are found over intervals. The total area under the curve must be 1.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

The Central Limit Theorem (CLT) states that, given a sufficiently large sample size n, the sampling distribution of the sample mean will be approximately normal, regardless of the shape of the original population distribution.

Mathematically, if we take random samples of size n from a population with mean Î¼ and standard deviation Ïƒ, the sample mean X follows a normal distribution

Works for Any Population Distribution
Even if the population is skewed or non-normal, the sample means will be approximately normal for large n (usually nâ‰¥30).

Since sample means are normally distributed, we can use Z-scores and T-tests to make probability calculations.

CLT allows us to estimate population parameters and apply techniques like confidence intervals and hypothesis testing.

21.  What are the assumptions for applying the Central Limit Theorem?

The Central Limit Theorem (CLT) states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population distribution. However, certain assumptions must be met for CLT to hold effectively.

Random Sampling : The sample data must be randomly selected from the population. If the data is biased or non-random, CLT may not apply properly.

Sample Size Must Be Large Enough : A larger sample size makes the CLT more effective.Rule of Thumb: nâ‰¥30 (for moderately skewed distributions).

If the population is highly skewed, an even larger sample size is needed.
Independence of Observations : Each data point should be independent of the others. This means one observation should not influence another.

Finite Variance (No Extreme Outliers): The population should have finite variance (no infinite or undefined variance).Extreme outliers or a population with infinite variance (like a Cauchy distribution) can disrupt CLT.

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?

```python
def bernoulli_pmf(x, p):
    
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")
    
    return p if x == 1 else 1 - p


p_success = 0.7  
print(f"P(X=1): {bernoulli_pmf(1, p_success)}")
print(f"P(X=0): {bernoulli_pmf(0, p_success)}")
```

19. Implement hypothesis testing using Z-statistics for a sample dataset?

```python
import numpy as np
import scipy.stats as stats

def z_test(sample, population_mean, population_std, alpha=0.05):
    
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    
    
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    
    
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    if p_value < alpha:
        result = "Reject the null hypothesis (H0)"
    else:
        result = "Fail to reject the null hypothesis (H0)"
    
    
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z_score:.2f}")
    print(f"P-value: {p_value:.5f}")
    print(f"Conclusion: {result}")


np.random.seed(42)  
sample_data = np.random.normal(loc=72, scale=8, size=30)  


population_mean = 70
population_std = 10


z_test(sample_data, population_mean, population_std)
```

20. Create a confidence interval for a dataset using Python and interpret the result

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  
    sample_size = len(data)
    
    z_critical = stats.norm.ppf((1 + confidence) / 2) 
    
    
    margin_of_error = z_critical * (sample_std / np.sqrt(sample_size))
    
    
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Confidence Interval ({int(confidence*100)}%): [{lower_bound:.2f}, {upper_bound:.2f}]")


np.random.seed(42)
data_sample = np.random.normal(loc=70, scale=10, size=30) 


confidence_interval(data_sample, confidence=0.95)
```

27. Implement Z-score normalization on a dataset and visualize its transformation

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


np.random.seed(42)
mean, std_dev, size = 50, 15, 1000  
dataset = np.random.normal(mean, std_dev, size)


z_scores = (dataset - np.mean(dataset)) / np.std(dataset)


plt.figure(figsize=(12, 5))


plt.subplot(1, 2, 1)
sns.histplot(dataset, bins=30, color='blue', kde=True)
plt.title("Original Dataset")
plt.xlabel("Values")
plt.ylabel("Frequency")


plt.subplot(1, 2, 2)
sns.histplot(z_scores, bins=30, color='red', kde=True)
plt.title("Z-Score Normalized Dataset")
plt.xlabel("Z-Scores")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()


print(f"Original Mean: {np.mean(dataset):.2f}, Original Std Dev: {np.std(dataset):.2f}")
print(f"Normalized Mean: {np.mean(z_scores):.2f}, Normalized Std Dev: {np.std(z_scores):.2f}")
```

28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

def analyze_distribution(size=1000, mean=50, std_dev=10):
   
    np.random.seed(42)
    data = np.random.normal(mean, std_dev, size)

  
    skewness = skew(data)
    kurt = kurtosis(data, fisher=True) 

    # Print results
    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis (Excess): {kurt:.4f}")
    print(f"Total Kurtosis: {kurt + 3:.4f}") 
    
 
    plt.figure(figsize=(8, 5))
    sns.histplot(data, bins=30, kde=True, color='skyblue')
    plt.title("Histogram & KDE of Normally Distributed Data")
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.axvline(np.mean(data), color='red', linestyle='dashed', label="Mean")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()


analyze_distribution()
```
